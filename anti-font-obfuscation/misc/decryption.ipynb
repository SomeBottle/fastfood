```python
from xml.etree import ElementTree as XMLTree
```

```python
tree = XMLTree.parse('./exam.xml')
root=tree.getroot()
```

```python
# 将元素的所有孩子元素的 raw_xml 拼接在一起
def get_raw_children(element):
    return b''.join([XMLTree.tostring(child,encoding='UTF-8') for child in element])
```

```python
glyphs_to_modified_uni={}
modified_uni_to_glyphs={}

for ttgl in root.iter('TTGlyph'):
    name=ttgl.get('name')
    if name.startswith('uni'):
        # 找到 Unicode 编码的字形，记录相应的 XML 元素
        raw_xml=get_raw_children(ttgl)
        # 建立字形和加密字体中 unicode 的映射
        glyphs_to_modified_uni[raw_xml]=name
        modified_uni_to_glyphs[name]=raw_xml
```

原字体为 Source Han Sans SC VF ExtraLight  
https://github.com/adobe-fonts/source-han-sans/releases/tag/2.004R

```python
original_tree=XMLTree.parse('./original.xml')
original_root=original_tree.getroot()
```

```python
modified_u_to_original_u={}
for ttgl in original_root.iter('TTGlyph'):
    name=ttgl.get('name')
    if name.startswith('uni'):
        raw_xml=get_raw_children(ttgl)
        # 建立混淆后字形 unicode 到 原字形 unicode 的映射
        if raw_xml in glyphs_to_modified_uni:
            # 找到了匹配的字形
            modified_u_to_original_u[glyphs_to_modified_uni[raw_xml]]=name
```
